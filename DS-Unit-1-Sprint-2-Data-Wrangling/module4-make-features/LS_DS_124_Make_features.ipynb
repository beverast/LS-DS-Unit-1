## Work with strings

For machine learning, we usually want to replace strings with numbers.

We can get info about which columns have a datatype of "object" (strings)

```python
df.describe(exclude='number')
```

### Convert `int_rate`

Define a function to remove percent signs from strings and convert to floats

```python
df['int_rate'].str.strip('%').astype(float).head()
```

```python
def remove_percent(string: str) -> float:
    return float(string.strip('%'))

remove_percent('13.63%')
```

Apply the function to the `int_rate` column

```python
df['int_rate'] = df['int_rate'].apply(remove_percent)
```

```python
(df['int_rate'] / 100).head()
```

### Clean `emp_title`

Look at top 20 titles

```python
df['emp_title'].value_counts().head(20)
```

How often is `emp_title` null?

```python
df['emp_title'].isna().sum()
```

Clean the title and handle missing values

```python
import numpy as np


examples = ['owner', 'Supervisor ',
            ' Project Manager', np.NaN]

def clean_title(x):
    """Capitalize, strip spaces, replace NaN with 'Unknown' """
    if isinstance(x, str):
        return x.strip().title()
    else:
        return 'Unknown'

[clean_title(x) for x in examples]
```

```python
df['emp_title'].head(20), df['emp_title'].apply(clean_title).head(20)
```

```python
df['emp_title'] = df['emp_title'].apply(clean_title)
```

```python
df['emp_title'].value_counts().head(20)
```

### Create `emp_title_manager`

pandas documentation: [`str.contains`](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.str.contains.html)

```python
df['emp_title_manager'] = df['emp_title'].str.contains('Manager', regex=False)
```

## Work with dates

pandas documentation
- [to_datetime](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.to_datetime.html)
- [Time/Date Components](https://pandas.pydata.org/pandas-docs/stable/timeseries.html#time-date-components) "You can access these properties via the `.dt` accessor"

```python
# What is the type of this column?
df['issue_d'].head().values
```

```python
df['issue_d'].describe()
```

```python
df['issue_d'] = pd.to_datetime(df['issue_d'], infer_datetime_format=True)
```

```python
df['issue_d'].head().values
```

```python
# Describe now gives more useful information
# Notice that the dtype of the feature data is 'datetime64[ns]'
# and the dtype of the describe df is 'object'
df['issue_d'].describe()
```

```python
df['issue_year'] = df['issue_d'].dt.year
df['issue_month'] = df['issue_d'].dt.month
```

```python
df.shape
```

```python
df['issue_month'].sample(n=10).values
```

```python
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'], infer_datetime_format=True)
```

```python
df['earliest_cr_line'].head().values
```

```python
df['earliest_cr_line'].head()
```

```python
# Wow. So easy. 
df['days_from_earliest_cr_line_to_issue'] = (
    df['issue_d'] - df['earliest_cr_line']).dt.days
```

```python
df['days_from_earliest_cr_line_to_issue'].describe()
```

```python
df['days_from_earliest_cr_line_to_issue'].max() / 365
```

### How to convert multiple columns to datetime programmatically

```python
# Search for columns with '_d'
[col for col in df if col.endswith('_d')]
```

```python
for col in ['last_pymnt_d', 'next_pymnt_d', 'last_credit_pull_d']:
    df[col] = pd.to_datetime(df[col], infer_datetime_format=True)
```

```python
df.describe(include='datetime')
```

## LendingClub Stretch Goals

### Remove percent signs and convert to floats, and handle missing values.

```python
df['revol_util'] = df['revol_util'].str.strip('%').astype(float)
df['revol_util'] = df['revol_util'].fillna(df['revol_util'].mean())
```

```python
df['revol_util'].describe()
```

### Modify the emp_title column to replace titles with 'Other' if the title is not in the top 20.

```python
top_titles = df['emp_title'].value_counts().index.tolist()[:20]
df['emp_title'] = df['emp_title'].where(df['emp_title'].isin(top_titles), 'Other')
```
